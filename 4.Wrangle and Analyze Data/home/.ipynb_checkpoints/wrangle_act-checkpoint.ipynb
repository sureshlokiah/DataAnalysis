## 5. Fixing the dog names

#### Define

- Fix some of the invalid dog names with the name extraced from 'text' field

#### Code

```python
## First count the 'Invalid dog names'
df_invalid_name = df_twitter_clean[df_twitter_clean.name.isin(lowercase_names)]
print("Count of Invalid dog names : " + str(len(df_invalid_name)))
```

```python
## Iterate and fix names
pd.set_option('mode.chained_assignment', None)

for i, row in df_twitter_clean.iterrows():
    if 'name is' in row.text:
        match_obj = re.match(r'.*?name is (\w+).*', row.text)
        if match_obj:
            name = match_obj.group(1)
            df_twitter_clean.name[i] = name
    
    if 'named' in row.text:
        match_obj = re.match(r'.*?named (\w+).*', row.text)
        if match_obj:
            name = match_obj.group(1)
            df_twitter_clean.name[i] = name
```

```python
#### Test
```

## 9. Fix dog_breed

#### Define

- Dog breed is gauged from image_predictions confidence score.

#### Code

```python
# First create a new column 'dog_brreed'

df_image_predictions_clean['dog_breed'] = None

## Update it with value based on confidence score.
for i, row in df_image_predictions_clean.iterrows():
    breed = None
    conf = -1
    #compare p1 to p2
    if row.p1_dog:
        breed = row.p1
        conf = row.p1_conf
    if row.p2_dog:
        if row.p2_conf > conf:
            breed = row.p2
            conf = row.p2_conf
    if row.p3_dog:
        if row.p3_conf > conf:
            breed = row.p3
            conf = row.p3_conf
    ##
    if breed != None:
        # Replace _ to space
        breed = breed.replace("_", " ").title()
        
    ## Set the highest confidence dog breed value
    df_image_predictions_clean.dog_breed[i] = breed
    ## Replace

```
